# Q2: Planetary positions

```python
import numpy as np
import matplotlib.pyplot as plt


a = np.array([0.39, 0.72, 1.00, 1.52, 5.20, 9.54, 19.22, 30.06, 39.48])
P = np.array([0.24, 0.62, 1.00, 1.88, 11.86, 29.46, 84.01, 164.8, 248.09])
names = ["Mercury", "Venus", "Earth", "Mars", "Jupiter", "Saturn", 
         "Uranus", "Neptune", "Pluto"]

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(a, P, 'o')
ax.set_xlabel("distance (a.u.)")
ax.set_ylabel("period (y)")
ax.set_xscale("log")
ax.set_yscale("log")
#plt.text(a, P, names)

for i in range(len(a)):
    plt.text(a[i], P[i], names[i])
```

# Q3: A perfect circle

```python
import numpy as np
import matplotlib.pyplot as plt
import random

R = 1
theta = np.linspace(0, 2*np.pi, 100)
x = R*np.cos(theta)
y = R*np.sin(theta)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y, '.')
plt.fill(x, y, 'aqua')
ax.set_xlabel('x')
ax.set_label('y')
ax.set_aspect('equal')



color = ['blue', 'red', 'green', 'yellow', 'orange', 'aqua', 'purple', 'pink', 'brown', 'lavender']
def drew_circle(x0, y0, R):
    theta = np.linspace(0, 2*np.pi, 100)
    x = x0+R*np.cos(theta)
    y = y0+R*np.sin(theta)
    plt.fill(x, y, random.choice(color))
    ax.set_aspect('equal')
    
fig = plt.figure()
ax = fig.add_subplot(111)

alpha = np.linspace(0, 2*np.pi, 10) 
x0 = np.cos(alpha)
y0 = np.sin(alpha)
R = np.linspace(0.1, 0.4, 10)

for i in range(len(x0)):
    drew_circle(random.choice(x0), random.choice(y0), random.choice(R))   
```

# Q4: Climate

```python
import numpy as np
import matplotlib.pyplot as plt

y, T, sT = np.loadtxt('nasa-giss.txt', skiprows=8, unpack = 'True')

fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(y, sT, label='smoothed T')
ax.set_xlabel('Year')
ax.set_ylabel('Temperature')
plt.legend()
color = np.where(T<0,'blue', 'red')
ax.scatter(y, T,  c = color)

plt.show()

  
```

# Q5: Subplots

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 500)
f = np.sin(x)**2+(1/4)*x*np.cos(10*x)

dx = x[1] - x[0]
f1 = (f[1:] - f[:-1])/dx

f2 = (f1[1:] - f1[:-1])/dx

print(len(x), len(f1), len(f2))
fig, axs = plt.subplots(3)
axs[0].plot(x, f)
axs[1].plot(x[1:], f1)
axs[2].plot(x[2:], f2)

axs[0].set_ylabel('f')
axs[1].set_ylabel('f1')
axs[2].set_ylabel('f2')
axs[2].set_xlabel('x')
```
